````python
import random

# Файлдарды окуу функциясы
def read_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        return set(file.read().splitlines())

# ru_kg_udahin.txt файлын ачуу
def get_random_entries(ru_kg_filepath, exclude_filepath, num_entries=100):
    # Окуу 12000.txt файлы жана ru_kg_udahin.txt файлы
    exclude_words = read_file(exclude_filepath)
    all_entries = list(read_file(ru_kg_filepath))

    # Чыпкалоо: 12000.txt файлына кирбеген сөздөрдү алуу
    filtered_entries = [entry for entry in all_entries if entry not in exclude_words]

    # Рандом түрдө 100 макала тандоо
    return random.sample(filtered_entries, num_entries)

# Файлдардын жолдорун көрсөтүү
ru_kg_filepath = 'ru_kg_udahin.txt'
exclude_filepath = '12000.txt'

# 100 сөздүк макаласын тандоо
random_entries = get_random_entries(ru_kg_filepath, exclude_filepath)

# Тандалган макалаларды көрсөтүү
for entry in random_entries:
    print(entry)
````

#### Loop through the gold corpus lines, feed the texts to fine-tuned Gemini Flash 1.5, get outputs, and save them into a file for further evaluation against the gold corpus outputs

````python
import json

output_file = 'gemini_flash_outputs.jsonl'


def save_results():
  with open(output_file, 'w', encoding='utf-8') as f:
      for item in gemini_flash_outputs:
          f.write(json.dumps(item, ensure_ascii=False) + '\n')

  print(f"Progress saved to {output_file}")
````

````python
from google.generativeai.types import HarmCategory, HarmBlockThreshold

# TODO: fix duplication error
processed_task_ids = {output['task-id'] for output in gemini_flash_outputs}

with open('gold_corpus_structured_yudakhin.jsonl', 'r') as f:
    for line in f:
        try:
            data = json.loads(line)
            task_id = data.get('task-id')

            # Skip if already processed
            if task_id in processed_task_ids:
                continue

            dict_entry = data.get('text')
            prompt = f"{system_message}\nDictionary entry:\n\n{dict_entry}"

            # Make the request to the model and handle rate limits
            try:
                result = tuned_model.generate_content(
                    prompt,
                    safety_settings={
                        HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_NONE,
                        HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_NONE,
                        HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_NONE,
                        HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_NONE
                    })
            except Exception as e:
                print(f"Error generating content for task-id {task_id}: {e}")
                time.sleep(60)  # Sleep for 1 minute and retry
                continue

            # Append result to the output
            gemini_flash_outputs.append({
                'task-id': task_id,
                'text': dict_entry,
                'json': result.text
            })

            # Save progress periodically to avoid data loss
            if len(gemini_flash_outputs) % 10 == 0:
                save_results()

        except Exception as e:
            print(f"General error for task-id {task_id}: {e}")
            time.sleep(60)  # Sleep for 1 minute and continue

# Final save after completion
save_results()
````

````python
import json

output_file = 'gemini_flash_outputs.jsonl'
deduped_output_file = 'gemini_flash_outputs_deduped.jsonl'

def dedupe_jsonl(input_file, output_file):
    unique_items = {}

    with open(input_file, 'r', encoding='utf-8') as f:
        for line in f:
            item = json.loads(line)
            task_id = item.get('task-id')
            if task_id not in unique_items:
                unique_items[task_id] = item

    with open(output_file, 'w', encoding='utf-8') as f_out:
        for item in unique_items.values():
            f_out.write(json.dumps(item, ensure_ascii=False) + '\n')

    print(f"Deduplicated data saved to {output_file}")

dedupe_jsonl(output_file, deduped_output_file)
````

````python
output_file = 'untrained_gemini_flash_outputs.jsonl'
untrained_gemini_flash_outputs = []

processed_task_ids = set()

try:
    with open(output_file, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            processed_task_ids.add(data['task-id'])
except FileNotFoundError:
    pass

def save_results():
    with open(output_file, 'w', encoding='utf-8') as f:
        for item in untrained_gemini_flash_outputs:
            f.write(json.dumps(item, ensure_ascii=False) + '\n')
    print(f"Progress saved to {output_file}")


gemini_model = genai.GenerativeModel('gemini-1.5-flash-001')
config = genai.GenerationConfig(temperature=0)

with open('gold_corpus_structured_yudakhin.jsonl', 'r') as f:
    for line in f:
        try:
            data = json.loads(line)
            task_id = data.get('task-id')

            if task_id in processed_task_ids:
                continue

            dict_entry = data.get('text')
            prompt = f"{system_message}\nDictionary entry:\n\n{dict_entry}"

            try:
                result = gemini_model.generate_content(
                    prompt,
                    generation_config=config,
                    safety_settings={
                        HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_NONE,
                        HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_NONE,
                        HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_NONE,
                        HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_NONE
                    }
                )

                if not hasattr(result, 'text') or not result.text:
                    print(f"Response blocked or invalid for task-id {task_id}. Skipping.")
                    continue  # Skip if the result is invalid

            except Exception as e:
                print(f"Error generating content for task-id {task_id}: {e}")
                time.sleep(60)  # Sleep for 1 minute and retry
                continue

            untrained_gemini_flash_outputs.append({
                'task-id': task_id,
                'text': dict_entry,
                'json': result.text
            })

            if len(untrained_gemini_flash_outputs) % 10 == 0:
                save_results()

        except Exception as e:
            print(f"General error for task-id {task_id}: {e}")
            time.sleep(60)  # Sleep for 1 minute and continue

# Final save after completion
save_results()
````

````python
def get_inference_with_few_shot_examples(output_file, system_prompt):
  outputs = []

  processed_task_ids = set()

  try:
      with open(output_file, 'r', encoding='utf-8') as f:
          for line in f:
              data = json.loads(line)
              processed_task_ids.add(data['task-id'])
  except FileNotFoundError:
      pass

  def save_results():
      with open(output_file, 'w', encoding='utf-8') as f:
          for item in outputs:
              f.write(json.dumps(item, ensure_ascii=False) + '\n')
      print(f"Progress saved to {output_file}")

  gemini_model = genai.GenerativeModel('gemini-1.5-flash-001')
  config = genai.GenerationConfig(temperature=0)

  with open('gold_corpus_structured_yudakhin.jsonl', 'r') as f:
      for line in f:
          try:
              data = json.loads(line)
              task_id = data.get('task-id')

              if task_id in processed_task_ids:
                  continue

              dict_entry = data.get('text')
              prompt = f"{system_prompt}\nEntry:\n\n{dict_entry}\nComputer-readable format:"

              try:
                  result = gemini_model.generate_content(
                      prompt,
                      generation_config=config,
                      safety_settings={
                          HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_NONE,
                          HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_NONE,
                          HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_NONE,
                          HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_NONE
                      }
                  )

                  if not hasattr(result, 'text') or not result.text:
                      print(f"Response blocked or invalid for task-id {task_id}. Skipping.")
                      continue  # Skip if the result is invalid

              except Exception as e:
                  print(f"Error generating content for task-id {task_id}: {e}")
                  time.sleep(60)  # Sleep for 1 minute and retry
                  continue

              outputs.append({
                  'task-id': task_id,
                  'text': dict_entry,
                  'json': result.text
              })

              if len(outputs) % 10 == 0:
                  save_results()

          except Exception as e:
              print(f"General error for task-id {task_id}: {e}")
              time.sleep(60)  # Sleep for 1 minute and continue
  # Final save after finishing
  save_results()
  # End
````

````python
import time
from google.colab import userdata
from openai import OpenAI

SLEEP_TIME = 0.3

client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))


def get_inference_with_few_shot_examples_from_open_ai(output_file, system_prompt, model_name='gpt-4o-2024-08-06'):
  outputs = []

  processed_task_ids = set()

  try:
      with open(output_file, 'r', encoding='utf-8') as f:
          for line in f:
              data = json.loads(line)
              processed_task_ids.add(data['task-id'])
  except FileNotFoundError:
      pass

  def save_results():
      with open(output_file, 'w', encoding='utf-8') as f:
          for item in outputs:
              f.write(json.dumps(item, ensure_ascii=False) + '\n')
      print(f"Progress saved to {output_file}")

  with open('gold_corpus_structured_yudakhin.jsonl', 'r') as f:
      for line in f:
          try:
              data = json.loads(line)
              task_id = data.get('task-id')

              if task_id in processed_task_ids:
                  continue

              dict_entry = data.get('text')
              try:
                  response = client.chat.completions.create(
                      model=model_name,
                      response_format={ "type": "json_object" }, # JSON is being returned!
                      messages=[
                          {"role": "system", "content": system_prompt},
                          {"role": "user", "content": dict_entry}
                      ],
                      temperature=0,
                      max_tokens=4096,
                      top_p=1
                  )
              except Exception as e:
                  print(f'Exception processing entry "{dict_entry}": {e}')
                  continue

              outputs.append({
                  'task-id': task_id,
                  'text': dict_entry,
                  'json': response.choices[0].message.content
              })

              if len(outputs) % 10 == 0:
                  save_results()

          except Exception as e:
              print(f"General error for task-id {task_id}: {e}")
              time.sleep(60)  # Sleep for 1 minute and continue
  # Final save after finishing
  save_results()
  # End
````
